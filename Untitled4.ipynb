#1.

Matrices (n, k) and (k, m). In the matrix multiplication, the number of multiplications will be k * m * n and the number of additions will be (k - 1) * m * n.

Explanation: Let the two matrices be A and B and C = A x B.
Then, each term of C, C<sub>ij</sub>, is the result of the scalar product of each the ith row of A and the jth column of B. There will be k multiplications here. Since, the size of C will be (n, m), the total number of terms will be n * m. Thus, the total number of multiplications will be n * m * k.

Since, there will be k terms in each row of A and each column of B, in calculating a term C<sub>ij</sub>, we will have to add k terms together. Thus, k-1 additions.

```python
import time
import random
import numpy as np
```

```python
#matrix multiplication using list of lists

n = int(input("n = "))
k = int(input("k = "))
m = int(input("m = "))

def generate_matrix (rows, cols):
  matrix = [[random.randint(0, 10) for _ in range (cols)] for _ in range (rows)]
  return matrix

def multiply_matrices(A, B, n, k, m):
  C = []
  for i in range (n):
    row = []
    for j in range (m):
      cell_sum = 0
      for l in range (k):
        cell_sum += A[i][l] * B[l][j]
      row.append(cell_sum)
    C.append(row)
  return C

print("Matrix A")
A = generate_matrix(n, k)
for row in A:
  print(row)
print("Matrix B")
B = generate_matrix(k, m)
for row in B:
  print(row)
print("The product of matrices A and B is")
start = time.time()
C = multiply_matrices(A, B, n, k, m)
end = time.time()
print("Using List of Lists")
for row in C:
  print(row)
print(end - start)

#matrix multiplication using numpy arrays
print("Using numpy arrays")
A = np.array(A)
B = np.array(B)
start = time.time()
C = np.dot(A, B)
end = time.time()
print(C)
print(end - start)
```

Numpy is faster than lists.
This is because:
1. Numpy arrays store data in continugous memory addresses unlike python lists where the data can be spread out throughout the memory.
2. Numpy uses vectorisation which means that an operation can be performed on the entire array at once unlike lists where iterating through the complete list is necessary.
3. Due to the fixed data type, Numpy arrays use less memory than lists. This also results in faster processing.

Using this method, the time complexity of finding the median of an array is O(n<sup>2</sup>).

A better way to find the median would be to sort the array and find the middle term(s). In this case, the time complexity would be O(nlogn).

```python
def naive_median(arr):
  arr_copy = list(arr)
  n = len(arr)
  for _ in range(n//2):
    max_arr = float('-inf')
    for elt in arr_copy:
      if elt > max_arr:
        max_arr = elt
    arr_copy.remove(max_arr)
  next_max = float('-inf')
  for elt in arr_copy:
    if elt > next_max:
      next_max = elt
  median = (next_max + max_arr)/2
  if n % 2 == 0:
    median = (max_arr + next_max)/2
  else:
    median = next_max
  return median

def sorted_median(arr):
  sorted_arr = sorted(arr)
  n = len(arr)
  if n%2 == 0:
    median = (sorted_arr[n//2 - 1] + sorted_arr[n//2])/2
  else:
    median = sorted_arr[n//2]
  return median
n = int(input())
a = [random.randint(0, 10) for i in range (n)]
print(a)
start = time.time()
print(naive_median(a))
end = time.time()
print("Time for naive median:", end - start)
start = time.time()
print(sorted_median(a))
end = time.time()
print("Time for sorted median", end - start)
a_np = np.array(a)
start = time.time()
print(np.median(a_np))
end = time.time()
print("Time for numpy array:", end - start)
```

For small n, numpy is slow. For example, even at n = 100, it's slower than both naive median and sorted median. At n = 1000, it's faster than naive median but slower than sorted median. At n = 10000, it's faster than both naive median and sorted median. Thus, at larger values of n, numpy is the fastest.

```python
import jax.numpy as jnp
from jax import grad
```

```python
def f(x, y):
  return jnp.sum(x*x*y + y*y*y*jnp.sin(x))

# my analytical solution for gradient with respect to x
def x_gradient(x, y):
  return 2*x*y + y**3*np.cos(x)
# my analytical solution for gradient with respect to y
def y_gradient(x, y):
  return x*x + 3*y*y*np.sin(x)

x_grad = grad(f, argnums = 0)
y_grad = grad(f, argnums = 1)
n = 10
x_arr = np.float32(np.random.randint(0, 10, size = (n,)))
y_arr = np.float32(np.random.randint(0, 10, size = (n,)))
print("gradient w.r.t x at random values from Jax")
print(x_grad(x_arr, y_arr))
print("gradient w.r.t x at random values according to analytical solution")
print(x_gradient(x_arr, y_arr))
print("gradient w.r.t y at random values from Jax")
print(y_grad(x_arr, y_arr))
print("gradient w.r.t y at random values according to analytical solution")
print(y_gradient(x_arr, y_arr))
```

It can be seen that the gradients on each point is the same in both the Jax and the analytical solutions.

```python
class Year:
  def __init__(self, year, branches):
    self.year = year
    self.branches = branches

class Branch:
  def __init__(self, branch, students):
    self.branch = branch
    self.students = students

class Student:
  def __init__(self, roll_no, name, marks):
    self.roll_no = roll_no
    self.name = name
    self.marks = marks

class Marks:
  def __init__(self, subject, marks):
    self.subject = subject
    self.marks = marks

Aarav = Student(1, "Aarav", [Marks("Maths", 90), Marks("Physics", 90), Marks("Chemistry", 90)])
Ajinkya = Student(21, "Ajinkya", [Marks("Maths", 85), Marks("Physics", 85), Marks("Chemistry", 85)])
Anvit =  Student(48, "Anvit", [Marks("Maths", 100), Marks("Physics", 100), Marks("Chemistry", 100)])
branch1_2024 = Branch("Electrical", [Aarav, Ajinkya])
branch2_2024 = Branch("CSE", [Anvit])
year_2024 = Year(2024, [branch1_2024, branch2_2024])
Aniket = Student(30, "Aniket", [Marks("Maths", 90), Marks("English", 95)])
branch1_2023 = Branch("CSE", [Aniket])
year_2023 = Year(2023, [branch1_2023])
student_database = [year_2024, year_2023]
```

```python
for year in student_database:
  print(year.year)
  for branch in year.branches:
      print("Branch:", branch.branch)
      for student in branch.students:
          print("  Roll No:", student.roll_no, "Name:", student.name)
          for marks in student.marks:
              print("    ", marks.subject, ":", marks.marks)
```

```python
def argmin(arr, axis = None):
  if axis is None:
    argmin = 0
    min_val = np.inf
    flattened_arr = np.flatten(arr, -1)
    for i in range(flattened_arr.size):
      if flattened_arr[i] < min_val:
        min_val = flattened_arr[i]
        argmin = i
    return argmin
  elif axis >= len(arr.shape):
    raise ValueError
  else:
    moved = np.moveaxis(arr, axis, 0)
    shape = moved.shape
    out_shape = shape[1:]
    result = np.empty(out_shape, dtype = int)
    for idx in np.ndindex(out_shape):
      min_idx = 0
      min_val = moved[0][idx]
      for i in range(1, shape[0]):
        if moved[i][idx] < min_val:
          min_val = moved[i][idx]
          min_idx = i
      result[idx] = min_idx
    return result
```
